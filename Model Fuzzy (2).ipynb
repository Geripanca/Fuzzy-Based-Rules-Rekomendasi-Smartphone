5. Membuat Logika Fuzzy Untuk Harga

```python
def harga_murah(harga):
    if harga <= 2500000:
        return 1
    elif harga > 2500000 and harga < 5000000:
        return (5000000 - harga) / (5000000 - 2500000)
    elif harga >= 5000000:
        return 0

def harga_normal(harga):
    if harga <= 2500000 or harga >= 20000000:
        return 0
    elif harga > 2500000 and harga <= 5000000:
        return (harga - 2500000) / (5000000 - 2500000)
    elif harga > 5000000 and harga < 15000000:
        return 1
    elif harga >= 15000000 and harga <= 20000000:
        return (20000000 - harga) / (20000000 - 15000000)

def harga_mahal(harga):
    if harga <= 15000000:
        return 0
    elif harga > 15000000 and harga < 20000000:
        return (harga - 15000000) / (20000000 - 15000000)
    elif harga >= 20000000:
        return 1

def predict_Harga(harga):
    try:
        murah = harga_murah(float(harga))  
        normal = harga_normal(float(harga))
        mahal = harga_mahal(float(harga))
        return murah, normal, mahal
    except Exception as e:
        print(f"Error predicting harga for {harga}: {e}")
        return None, None, None
```

```python
sql.execute("SELECT merek,harga FROM tb_datahp")

table_harga = BeautifulTable()

header_harga = ["No","merek","harga","murah","normal","mahal"]
table_harga.column_headers = header_harga
harga = 1
for row in sql.fetchall() :
    murah = predict_Harga(row["harga"])[0]
    normal = predict_Harga(row["harga"])[1]
    mahal = predict_Harga(row["harga"])[2]
    
    table_harga.append_row([harga, row["merek"], row["harga"], murah, normal, mahal])  
    harga = harga + 1  
print(table_harga)
```

6. Membuat Logika Fuzzy untuk RAM 

```python
def ram_kecil(RAM):
    if RAM <= 3:
        return 1
    elif 3 < RAM <= 6:
        return (6 - RAM) / (6 - 3)
    else:
        return 0

def ram_sedang(RAM):
    if RAM <= 3 or RAM >= 12:
        return 0
    elif 3 < RAM <= 6:
        return (RAM - 3) / (6 - 3)
    elif 6 < RAM <= 10:
        return 1
    elif 10 < RAM <= 12:
        return (12 - RAM) / (12 - 10)

def ram_besar(RAM):
    if RAM <= 10:
        return 0
    elif 10 < RAM <= 12:
        return (RAM - 10) / (12 - 10)
    else:
        return 1

def predict_ram(RAM):
    try:
        kecil = ram_kecil(float(RAM))  
        sedang = ram_sedang(float(RAM))
        besar = ram_besar(float(RAM))
        return kecil, sedang, besar
    except Exception as e:
        print(f"Error predicting harga for {RAM}: {e}")
        return None, None, None
```

```python
sql.execute("SELECT merek,RAM FROM tb_datahp")

table_RAM = BeautifulTable()

header_RAM = ["No","merek","RAM","kecil","sedang","besar"]
table_RAM.column_headers = header_RAM
RAM= 1
for row in sql.fetchall() :
    kecil = predict_ram(row["RAM"])[0]
    sedang = predict_ram(row["RAM"])[1]
    besar = predict_ram(row["RAM"])[2]
    table_RAM.append_row([harga, row["merek"], row["RAM"], kecil,sedang,besar])  
    RAM = RAM + 1  
print(table_RAM)
```

9. Fuzzyfikasi untuk Baterai

```python
def baterai_kecil(baterai):
    if baterai <= 2500:
        return 1
    elif 2500 < baterai <= 3000:
        return (3000 - baterai) / (3000 - 2500)
    else:
        return 0

def baterai_sedang(baterai):
    if baterai <= 2500 or baterai >= 5500:
        return 0
    elif 2500 < baterai <= 3000:
        return (baterai - 2500) / (3000 - 2500)
    elif 3000 < baterai <= 5000:
        return 1
    elif 5000 < baterai <= 5500:
        return (5500 - baterai) / (5500 - 5000)

def baterai_besar(baterai):
    if baterai <= 5000:
        return 0
    elif 5000 < baterai <= 5500:
        return (baterai - 5000) / (5500 - 5000)
    elif baterai >= 5500:
        return 1
def predict_baterai(baterai):
    try:
        kecil = baterai_kecil(float(baterai))  
        sedang = baterai_sedang(float(baterai))
        besar = baterai_besar(float(baterai))
        return kecil,sedang,besar
    except Exception as e:
        print(f"Error predicting harga for {baterai}: {e}")
        return None, None, None
```

10. Fuzzyfikasi untuk Kecepatan Prosesor

```python
def kecprosesor_lambat(kecprosesor):
    if kecprosesor <= 1:
        return 1
    elif 1 < kecprosesor <= 1.3:
        return (1.3 - kecprosesor) / (1.3 - 1)
    else:
        return 0

def kecprosesor_sedang(kecprosesor):
    if kecprosesor <= 1 or kecprosesor >= 3:
        return 0
    elif 1 < kecprosesor <= 1.3:
        return (kecprosesor - 1) / (1.3 - 1)
    elif 1.3 < kecprosesor <= 2:
        return 1
    elif 2 < kecprosesor <= 2.5:
        return (2.5 - kecprosesor) / (2.5 - 2)

def kecprosesor_cepat(kecprosesor):
    if kecprosesor <= 2:
        return 0
    elif 2 < kecprosesor <= 2.5:
        return (kecprosesor - 2) / (2.5 - 2)
    elif kecprosesor >= 2.5:
        return 1

def predict_kecprosesor(kecprosesor):
    try:
        lambat = kecprosesor_lambat(float(kecprosesor))  
        sedang = kecprosesor_sedang(float(kecprosesor))
        cepat = kecprosesor_cepat(float(kecprosesor))
        return lambat,sedang,cepat
    except Exception as e:
        print(f"Error predicting harga for {kecprosesor}: {e}")
        return None, None, None
```

```python
sql.execute("SELECT merek,kecprosesor FROM tb_datahp")

table_kecprosesor = BeautifulTable()

header_kecprosesor = ["No","merek","kecprosesor","lambat","sedang","cepat"]
table_kecprosesor.column_headers = header_kecprosesor
kecprosesor = 1
for row in sql.fetchall() :
    lambat = predict_kecprosesor(row["kecprosesor"])[0]
    sedang = predict_kecprosesor(row["kecprosesor"])[1]
    cepat = predict_kecprosesor(row["kecprosesor"])[2]
    
    table_kecprosesor.append_row([harga, row["merek"], row["kecprosesor"], lambat,sedang,cepat])  
    kecprosesor = kecprosesor + 1  
print(table_kecprosesor)
```

Defuzzyfikasi dengan Fungsi Max

```python
sql.execute("SELECT merek,harga FROM tb_datahp")

table_harga_sesuai = BeautifulTable()
table_harga_sesuai.column_headers = ["merek","harga","Defuzzyfikasi"]

for row in sql.fetchall() :
    murah, normal, mahal = predict_Harga(row["harga"])
    
    nilai_tertinggi = max(murah, normal, mahal)
    

    if nilai_tertinggi == murah:
        kategori_harga = "Murah"
    elif nilai_tertinggi == normal:
        kategori_harga = "Normal"
    elif nilai_tertinggi == mahal:
        kategori_harga = "Mahal"
    else:
        kategori_harga = "Tidak Diketahui"
    
    table_harga_sesuai.append_row([row["merek"], row["harga"], kategori_harga])

print(table_harga_sesuai)
```

Membuat Rules/Knowledge Misalnya Jika Harga murah dan kecepatan prosesor sedang atau cepat maka Smartphone akan direkomendasikan

```python
sql.execute("SELECT merek, harga, kecprosesor FROM tb_datahp")

table_rekomendasi = BeautifulTable()
table_rekomendasi.column_headers = ["merek", "harga", "kecprosesor", "Rekomendasi"]

for row in sql.fetchall():
    # Mendapatkan nilai keanggotaan dari kedua variabel
    murah, normal, mahal = predict_Harga(row["harga"])
    lambat, sedang, cepat = predict_kecprosesor(row["kecprosesor"])

    # Mengambil nilai keanggotaan tertinggi dari masing-masing variabel
    nilai_tertinggi_harga = max(murah, normal, mahal)
    nilai_tertinggi_kecprosesor = max(lambat, sedang, cepat)

    # Menentukan kategori rekomendasi
    if nilai_tertinggi_harga == murah and (nilai_tertinggi_kecprosesor == cepat or nilai_tertinggi_kecprosesor == sedang):
        rekomendasi = "Rekomendasi"
    else:
        rekomendasi = "Tidak Direkomendasikan"

    table_rekomendasi.append_row([row["merek"], row["harga"], row["kecprosesor"], rekomendasi])

print(table_rekomendasi)
```
